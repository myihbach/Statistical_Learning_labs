# **2 - SPLITTAGE DES DONNEES**



```
from sklearn.model_selection import train_test_split
```

```
# splittage des donnes , 75 %  pour l'entrainement , 25 % pour le test ( 75 / 25 est les valeurs par  defaut de la fonction train_test_split )
X_train, X_test, y_train, y_test = train_test_split(irisdataset['data'],irisdataset['target'])
print("Shape of X_train : {}".format(X_train.shape))
print("Shape of X_test : {}".format(X_test.shape))
print("Shape of Y_train : {}".format(y_train.shape))
print("Shape of Y_test : {}".format(y_test.shape))

```

# **4 - Classification par lalgorithme des K voisins (K=1)**

```
from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

### Prediction pour une nouvelle entre X_new qui n'appartient pas aux donnes d'entrainement

```
def iris_type(a):
  if a == 0 :
    return "Setosa"
  elif a == 1 :
    return "Versicolor"
  elif a == 2 :
    return "Virginica"
  else : 
    return "No class is matched"
```

```
X_new = np.array([[5, 2.9, 1, 0.2]]) 
y_pred = knn.predict(X_new)
print("La prediction de X_new : {}".format(iris_type(y_pred)))
```

# **5 - EVALUATION DES PERFORMANCES DU CLASSIFIEUR**

```
#mesure de l'efficacit du modle et Estimation de lerreur de prvision sur l'chantillon test 
from sklearn.metrics import accuracy_score
print("L'efficacit du modle est de : {} %".format(np.round(knn.score(X_test, y_test) * 100 , 2)))
```
